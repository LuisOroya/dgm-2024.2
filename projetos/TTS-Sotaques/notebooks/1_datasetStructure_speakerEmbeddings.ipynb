````python
import pandas as pd

metadata = pd.read_csv('/content/metadata_test_final.csv')
metadata = metadata[metadata['file_path'].str.contains("RE_EF")]
metadata['filename'] = metadata['file_path'].str.split('/').str[-1].str[:-4]
````

````python
import os
import shutil
````

````python
source_dir = "/content/dataset_RE/NURC_RE_EF"
destination_root_dir = "/content/dataset_RE/wav48"
#SR original não é 48kHz, mas esse precisa ser o nome do folder pra usar o formatter do vctk

for subdir in os.listdir(source_dir): # Para cada pasta de falante

  # Mover a pasta com wavs
  subdir_path = os.path.join(source_dir, subdir)
  destination_dir = os.path.join(destination_root_dir, subdir)
  shutil.move(subdir_path, destination_dir)

  # Criar pasta de textos do falante
  os.mkdir(os.path.join("/content/dataset_RE/txt", subdir))
````

````python
#!rm -r /content/dataset_RE/NURC_RE_EF
````

````python
txt_root_output_directory = "/content/dataset_RE/txt"
for speaker_id in os.listdir(txt_root_output_directory):
  speaker_path = os.path.join(txt_root_output_directory, speaker_id)

  for index, row in metadata.iterrows():

    if speaker_id in row['file_path']:
      filename_txt = f"{row['filename']}.txt"
      text = row['text']
      file_path = os.path.join(speaker_path, filename_txt)
      with open(file_path, 'w') as file:
          file.write(text)
````

````python
# Para treinamento multilíngue (neste caso, multisotaque), são requeridos datasets distintos,
# sendo cada um deles um objeto da classe BaseDatasetConfig

from TTS.config.shared_configs import BaseDatasetConfig

RE_config = BaseDatasetConfig(
    formatter="vctk_old", ### A estruturação do dataset é condicionada pelo formatter usado
    dataset_name="RE_EF",
    meta_file_train="",
    meta_file_val="",
    path="/content/dataset_RE",
    language="re",
)
````

````python
# Esta célula carrega o modelo que desemepenha o papel de encoder de falante e
# processa o(s) dataset(s) usando a função (compute_embeddings)

# OBS a saída da execução mostra que há também uma chamada ao Audio Processor, mas
# até o momento de criação deste notebook não identificamos exatamente exatamente qual
# parte do código interno do Coqui faz isso. É interessante controlar a chamada ao
# audio processor a fim de tornar possível ajustar suas configurações.

from TTS.bin.compute_embeddings import compute_embeddings

SPEAKER_ENCODER_CHECKPOINT_PATH = (
    "https://github.com/coqui-ai/TTS/releases/download/speaker_encoder_model/model_se.pth.tar"
)

SPEAKER_ENCODER_CONFIG_PATH = "https://github.com/coqui-ai/TTS/releases/download/speaker_encoder_model/config_se.json"

D_VECTOR_FILES = []  # List of speaker embeddings/d-vectors to be used during the training

DATASETS_CONFIG_LIST = [RE_config]

for dataset_conf in DATASETS_CONFIG_LIST:
    # Check if the embeddings weren't already computed, if not compute it
    embeddings_file = os.path.join(dataset_conf.path, "speakers.pth")
    if not os.path.isfile(embeddings_file):
        print(f">>> Computing the speaker embeddings for the {dataset_conf.dataset_name} dataset")
        compute_embeddings(
            SPEAKER_ENCODER_CHECKPOINT_PATH,
            SPEAKER_ENCODER_CONFIG_PATH,
            embeddings_file,
            old_speakers_file=None,
            config_dataset_path=None,
            formatter_name=dataset_conf.formatter,
            dataset_name=dataset_conf.dataset_name,
            dataset_path=dataset_conf.path,
            meta_file_train=dataset_conf.meta_file_train,
            meta_file_val=dataset_conf.meta_file_val,
            disable_cuda=False,
            no_eval=False,
        )
    D_VECTOR_FILES.append(embeddings_file)
````
