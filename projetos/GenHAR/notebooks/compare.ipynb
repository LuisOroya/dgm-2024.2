```python
import pandas as pd
import os

# Carregar o CSV
file_path = os.path.join("../reports/results_all/metrics_csv/utlility_ml_metrics.csv")
df = pd.read_csv(file_path)

# Filtrar colunas relevantes
df_filtered = df[['classifier', 'dataset_type', 'dataset', 'generator', 'f1-score']]

# Criar uma tabela pivotada para cada classificador
for clf in df_filtered['classifier'].unique():
    df_clf = df_filtered[df_filtered['classifier'] == clf]
    
    # Pivotar com dataset como √≠ndice, dataset_type nas colunas, para cada gerador
    pivot_table = df_clf.pivot_table(
        index='dataset',
        columns=['generator', 'dataset_type'],
        values='f1-score',
        aggfunc='max'  # Buscar o valor m√°ximo de F1-score
    )
    
    # Imprimir a tabela formatada
    print(f"\nTabela com m√°ximos para classificador: {clf}")

    max_highlight = pivot_table.style.highlight_max(axis=1)
    display(max_highlight)  
```
